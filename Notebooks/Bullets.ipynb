
# Bullet Mortgage Cash Flow Model: Detailed Explanation

This section explains the core block of the bullet mortgage cash flow model. The explanation is provided in two parts: a formal quant finance explanation with LaTeX formulas and a simpler "Nitish Sir" Hinglish explanation.

---

## Part A: Formal Quant Finance Explanation

Consider the following code block:

```python
for t in range(1, periods):
    M[t,4] = rate * M[t-1,1]         # Interest: I_t = r * N_(t-1)
    M[t,3] = 0                       # No scheduled principal repayment for a bullet mortgage
    scheduled_outstanding = M[t-1,1] - M[t,3]
    M[t,2] = scheduled_outstanding * CPR  # Prepayment = CPR * scheduled outstanding
    M[t,1] = scheduled_outstanding - M[t,2] # New outstanding notional = previous outstanding - prepayment
    M[t,5] = M[t,4] + M[t,2] + M[t,3]       # Total installment = interest + prepayment + principal repayment
```

Let's break down each step:

### **Step 1: Interest Calculation**

- **Code:**  
  ```python
  M[t,4] = rate * M[t-1,1]
  ```
- **Formula:**  
  $I_t = r \times N_{t-1}$
- **Explanation:**  
  The interest for period $t$ is computed by multiplying the interest rate $r$ with the outstanding notional (or principal) from the previous period $N_{t-1}$. This represents the cost of borrowing for that period.

---

### **Step 2: Scheduled Principal Repayment**

- **Code:**  
  ```python
  M[t,3] = 0
  ```
- **Explanation:**  
  In a bullet mortgage, there is no scheduled principal repayment during the life of the loan (except at maturity). Thus, the scheduled principal repayment is set to $0$ for all periods until the final one.

---

### **Step 3: Calculate Scheduled Outstanding**

- **Code:**  
  ```python
  scheduled_outstanding = M[t-1,1] - M[t,3]
  ```
- **Explanation:**  
  The scheduled outstanding is calculated as the previous period's outstanding notional minus the scheduled principal repayment. Since the scheduled repayment is $0$, the scheduled outstanding is simply:
  
  $$ \text{Scheduled Outstanding} = N_{t-1} - 0 = N_{t-1} $$

---

### **Step 4: Prepayment Calculation**

- **Code:**  
  ```python
  M[t,2] = scheduled_outstanding * CPR
  ```
- **Formula:**  
  $$ \text{Prepayment}_t = \text{CPR} \times (\text{Scheduled Outstanding}) $$
- **Explanation:**  
  The extra principal repayment (prepayment) is computed based on the Conditional Prepayment Rate (CPR). This means a fraction of the scheduled outstanding balance is repaid as prepayment.

---

### **Step 5: Update Outstanding Notional**

- **Code:**  
  ```python
  M[t,1] = scheduled_outstanding - M[t,2]
  ```
- **Formula:**  
  $$ N_t = \text{Scheduled Outstanding} - \text{Prepayment}_t $$
- **Explanation:**  
  The new outstanding notional for period $t$ is updated by subtracting the prepayment from the scheduled outstanding. This reflects the reduction in the principal due to prepayment.

---

### **Step 6: Calculate Total Installment**

- **Code:**  
  ```python
  M[t,5] = M[t,4] + M[t,2] + M[t,3]
  ```
- **Formula:**  
  $$ \text{Installment}_t = I_t + \text{Prepayment}_t + \text{Scheduled Repayment}_t $$
- **Explanation:**  
  The total installment for period $t$ is the sum of the interest, the prepayment, and the scheduled principal repayment. In a bullet mortgage, since the scheduled repayment is $0$, the installment is just the interest plus the prepayment.

## Part B: Nitish Sir Hinglish Explanation

Below is the same explanation in a simpler, conversational style:

1. **Interest Calculation:**
   - **Code:** `M[t,4] = rate * M[t-1,1]`
   - **Explanation:**  
     "Pehle, hum pichle saal ke outstanding principal par interest calculate karte hain. Matlab,  
     $I_t = r \times N_{t-1}$,  
     jisse pata chalta hai ki is period mein kitna interest charge hoga."

2. **No Scheduled Principal Repayment:**
   - **Code:** `M[t,3] = 0`
   - **Explanation:**  
     "Bullet mortgage mein regular principal repayment nahi hota, isliye isko $0$ set kar dete hain."

3. **Scheduled Outstanding Calculation:**
   - **Code:** `scheduled_outstanding = M[t-1,1] - M[t,3]`
   - **Explanation:**  
     "Scheduled outstanding nikalne ke liye, hum pichle saal ka outstanding lete hain aur scheduled repayment ko subtract karte hain. Yahan, kyunki repayment $0$ hai, scheduled outstanding barabar hai $N_{t-1}$ ke."

4. **Prepayment Calculation:**
   - **Code:** `M[t,2] = scheduled_outstanding * CPR`
   - **Explanation:**  
     "Prepayment calculate karte hain using CPR. Matlab,  
     $$ \text{Prepayment}_t = \text{CPR} \times (\text{Scheduled Outstanding}) $$  
     yeh extra payment ka hissa hai."

5. **Update Outstanding Notional:**
   - **Code:** `M[t,1] = scheduled_outstanding - M[t,2]`
   - **Explanation:**  
     "Naya outstanding principal update hota hai prepayment subtract karke. Yani,  
     $$ N_t = \text{Scheduled Outstanding} - \text{Prepayment}_t $$  
     isse pata chalta hai kitna principal abhi bhi bacha hai."

6. **Total Installment Calculation:**
   - **Code:** `M[t,5] = M[t,4] + M[t,2] + M[t,3]`
   - **Explanation:**  
     "Total installment nikalte hain jo hota hai interest, prepayment aur scheduled repayment ka sum. Yani,  
     $$ \text{Installment}_t = I_t + \text{Prepayment}_t + \text{Scheduled Repayment}_t $$  
     jisse humein us period ka total cash outflow milta hai."

This Markdown cell provides a comprehensive explanation of the bullet mortgage cash flow model, including all the underlying formulas using LaTeX. It is a useful reference when studying mortgage-backed securities or preparing for a quant interview.


````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



def Bullet(rate,notonal,periods,CPR):
    # it returns a matrix M such that
    # M=  [0,       1,            2,           3,                4              5]
    # M = [t  notional(t)  prepayment(t)  notional_quote(t)  interest_(t)  installment(t)]
    # WARNING! here "rate" and "periods" are quite general, the choice of getting year/month/day.. steps, depends on the rate
    # that the function receives. So, it is necessary to pass the correct rate to the function
    M=np.zeros((periods+1,6))
    M[:,0]=np.arange(periods+1)# time
    M[0,1]=notonal
    for t in range(1,periods):
        M[t,4]=rate*M[t-1,1]
        M[t,3]=0
        Scheduled_outstanding=M[t-1,1]-M[t,3]
        M[t,2]=Scheduled_outstanding*CPR
        M[t,1]=Scheduled_outstanding-M[t,2]# notional(t) = notional(t-1) - (repayment + prepayment)
        M[t,5]=M[t,4]+M[t,2]+M[t,3]
    M[periods,4]=rate*M[periods-1,1]
    M[periods,3]=M[periods-1,1]
    M[periods,5]=M[t,4]+M[t,2]+M[t,3]
        
    return M

def main():
    notonal=1000000
    r=0.05
    T_end=30
    Lambda=0.01
    M=Bullet(r, notonal, T_end, Lambda)
    for i in range(0,T_end+1):
        print("Ti={0}, Notional={1:.0f}, Prepayment={2:.0f}, Notional Repayment={3:.0f}, Interest Rate={4:.0f}, Installment={5:.0f} ".format(M[i,0],M[i,1],M[i,2],M[i,3],M[i,4],M[i,5]))
    plt.figure(1)
    plt.title("notonal of Bullets")
    plt.plot(M[:,0],M[:,1],'-r')
    plt.grid()
    plt.xlabel('time')
    plt.ylabel('notional')
    
    
main()
    
    

    
````
